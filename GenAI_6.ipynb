```python
import torch
from diffusers import DiffusionPipeline
import matplotlib.pyplot as plt
```

```python
art_pipeline = DiffusionPipeline.from_pretrained("CompVis/ldm-text2im-large-256").to("cuda" if torch.cuda.is_available() else "cpu")
```

```python
# Function to create digital art from a given prompt
def create_artistic_image(description):
    generated_image = art_pipeline(description, guidance_scale=8.0).images[0]  # Experiment with different guidance scale
    return generated_image

# Example prompt
art_description = "A party in outer space with futuristic technology"
generated_image = create_artistic_image(art_description)
```

```python
# Show the resulting artwork
plt.figure(figsize=(6,6))
plt.imshow(generated_image)
plt.axis("off")  # Hide axes
plt.show()
```
