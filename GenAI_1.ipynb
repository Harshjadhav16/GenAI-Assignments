```python
import random
from collections import defaultdict

def build_markov_chain(text, n=2):
    """Builds a Markov Chain model from the input text."""
    words = text.split()
    markov_chain = defaultdict(list)

    for i in range(len(words) - n):
        key = tuple(words[i:i + n])
        next_word = words[i + n]
        markov_chain[key].append(next_word)

    return markov_chain

def generate_paragraphs(markov_chain, num_paragraphs=3, max_sentences=5, max_words=50):
    """Generates multiple paragraphs using the Markov Chain model."""
    paragraphs = []

    for _ in range(num_paragraphs):
        sentences = []
        for _ in range(max_sentences):
            seed = random.choice(list(markov_chain.keys()))
            sentence = list(seed)

            for _ in range(max_words - len(seed)):
                key = tuple(sentence[-len(seed):])
                if key in markov_chain:
                    next_word = random.choice(markov_chain[key])
                    sentence.append(next_word)
                else:
                    break

            sentences.append(' '.join(sentence) + '.')

        paragraphs.append('\n'.join(sentences))  # Ensure sentences appear on separate lines within paragraphs

    return '\n\n'.join(paragraphs)

if __name__ == "__main__":
    # The following lines were indented incorrectly, causing the IndentationError.
    # Corrected by aligning them with the opening quote.
    sample_text = "In a faraway kingdom, a brave knight embarked on a perilous quest to slay a fearsome dragon. Along his journey, he encountered treacherous obstacles and mythical creatures. With unwavering determination, the knight faced each challenge, proving his courage and skill. Finally, he reached the dragon's lair, where an epic battle ensued. In the end, the knight emerged victorious, saving the kingdom from the dragon's wrath."
    markov_model = build_markov_chain(sample_text, n=2)
    generated_text = generate_paragraphs(markov_model, num_paragraphs=2, max_sentences=15, max_words=200)
    print(generated_text)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture

# Step 1: Generate synthetic 2D data from a Gaussian Mixture Model
np.random.seed(42)

# Parameters for 2 Gaussian components
mean1 = [2, 3]
cov1 = [[1, 0.5], [0.5, 1]]
mean2 = [-3, -2]
cov2 = [[1, -0.8], [-0.8, 1]]

# Generate samples
n_samples = 500
data1 = np.random.multivariate_normal(mean1, cov1, size=n_samples // 2)
data2 = np.random.multivariate_normal(mean2, cov2, size=n_samples // 2)

# Combine the two sets of samples
data = np.vstack([data1, data2])

# Step 2: Fit the Gaussian Mixture Model to the data
gmm = GaussianMixture(n_components=2, covariance_type='full')
gmm.fit(data)

# Step 3: Generate new samples from the fitted GMM
generated_data, _ = gmm.sample(n_samples=100)

# Step 4: Plot the original and generated data for visualization
plt.figure(figsize=(8, 6))
plt.scatter(data[:, 0], data[:, 1], c='blue', label='Original Data', alpha=0.5)
plt.scatter(generated_data[:, 0], generated_data[:, 1], c='red', label='Generated Data', alpha=0.5)
plt.title('Original and Generated Data from GMM')
plt.legend()
plt.show()
```
